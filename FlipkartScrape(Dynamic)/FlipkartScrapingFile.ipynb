```python
def flipkart(soup):
    info = []
    prdct_name = [i.text for i in soup.select('div._4rR01T')]
    price = [int(i.text[1:].replace(',','')) for i in soup.select('div._30jeq3._1_WHN1')]
    rating = [float(i.text) for i in soup.select('div._3LWZlK')]
    totalRating = [int(i.text.split()[0].replace(',','')) for i in soup.select('span._2_R_DZ')]
    totalReviews = [int(i.text.split()[3].replace(',','')) for i in soup.select('span._2_R_DZ')]
    for r in soup.select('ul._1xgFaf'):
        info.append([i.text for i in r.find_all('li')])
    df = pd.DataFrame({'Product Name': prdct_name, 'Price (In Rupees)': price, 'Rating': rating, 'Total No. of Rating': totalRating
                 ,'Total No. of Reviews': totalReviews, 'Specification': info})
    return df.to_csv('FlipkartAppledata.csv')
```

```python
flipkart(soup)
```
