```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.serialupdates.me/shark-tank-india-investors-names-sony-tv-new-show-entrepreneurs-list/'
```

```python
html = requests.get(url).text
html
```

```python
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
table = soup.find_all('table')[2]
table
```

```python
table.find_all('th')[0].text
```

```python
thead = [i.text for i in table.find_all('th')]
thead
```

```python
table.find_all('tbody')[0]
```

```python
table.find_all('tbody')[0].tr
```

```python
data = []
for r in table.find_all('tbody')[0]:
    data.append([i.text for i in r.find_all('td')])
```

```python
def sharktankindia(soup):
    data = []
    thead = [i.text for i in table.find_all('th')]
    for r in table.find_all('tbody')[0]:
        data.append([i.text for i in r.find_all('td')])
    df = pd.DataFrame(data, columns = thead).set_index('Sr No.')
    return df.to_csv('SharkTankIndia.csv')
```

```python
sharktankindia(soup)
```
